```python
import re
import requests
from bs4 import BeautifulSoup
import string
```

```python
def clearNames(names):
    result = re.search('Next page ((.*))', names)
    test = result.group(1)
    test2 = test.replace('(','').replace(')','').replace(' ','+')
    return test2
```

```python
def writeToFile(all_names):
    wordsDictionary = open("wordsDictionary.txt", "a")

    for a in all_names:
        wordsDictionary.write(a.encode('ascii', errors='ignore').decode() + " \n")
        print(a)
        
    wordsDictionary.close()
```

```python
nextPage = ''
for x in range(0,13):
    if x > 0:
        page = requests.get("https://sealonline.fandom.com/wiki/Special:AllPages?from="+nextPage)
        soup = BeautifulSoup(page.content, "html.parser")
        all_names = []
        for element in soup.select('a'):
            all_names.append(element.text)
        
        scrapped_names = []       
        try:
            for x in range(63,408):
                scrapped_names.append(all_names[x])
            nextPage = clearNames(all_names[62])
            writeToFile(scrapped_names)
        except IndexError:
            print('OUT OF INDEX')
            writeToFile(scrapped_names)
    else:
        page = requests.get("https://sealonline.fandom.com/wiki/Special:AllPages")
        
        soup = BeautifulSoup(page.content, "html.parser")
        all_names = []
        for element in soup.select('a'):
            all_names.append(element.text)
        
        scrapped_names = []       
        
        for x in range(62,407):
            scrapped_names.append(all_names[x])
                            
        nextPage = clearNames(all_names[61])
        writeToFile(scrapped_names)
    
```

```python
def writeToNewFile(all_names,file_names):
    wordsDictionary = open(file_names+".txt", "a")
    wordsDictionary.write(all_names + " \n")
        
    wordsDictionary.close()
```

```python
###############################Cleaning Data####################################
def cleaningData(string_input):
    str1 = string_input
    words = str1.split()
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in words]
    result = []
    curr_str = ""
    for item in stripped:
        curr_str += item
    return curr_str.lower()
```

```python
wordsDictionary = open("wordsDictionary.txt", "r")
for a in wordsDictionary:
    cleaned_str = cleaningData(a)
    writeToNewFile(cleaned_str,"cleanedWordsDictionary")
    
wordsDictionary.close()
```
